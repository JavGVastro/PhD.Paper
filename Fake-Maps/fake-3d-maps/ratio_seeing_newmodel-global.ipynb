```python
widths = [1, 2, 4, 8, 16, 32]
r_model = np.linspace(0.01,200,1000)
```

```python
#[rat,r,rat0,apparent_r0,true_r0]
sigE_0 = json.load(open("ratios_seeing_dens_cte.json"))
sigE_1 = json.load(open("ratios_seeing_sigE_1.json"))
sigE_2 = json.load(open("ratios_seeing_sigE_2.json"))
```

```python
r = np.array(sigE_0['1'][1])
```

```python
a = sigE_0 
R_NORM = 2.6
b_norms = []


fig, ax = plt.subplots(figsize=(8, 5))

colors = cmr.take_cmap_colors(
    "cmr.dusk",
    len(widths),
    cmap_range=(0.25, 0.95),
)



for i,c in zip(range(6), colors):
    rat0 = np.interp(R_NORM * widths[i], r, np.array(a[str(widths[i])][0]))
    b_norms.append(rat0)
    ax.plot(r/(R_NORM * widths[i]) , np.array(a[str(widths[i])][0])/rat0, color=c)
    
    #ax.plot(2 * widths[i]/(2 * widths[i]), a[str(widths[i])][2]/np.array(a[str(widths[i])][2]), marker="o",  color=c)
  
    
ax.plot(r_model , 2*(1 + (1/r_model)**2 )**-1, color='lime', linestyle="-", alpha = 0.7)

  
ax.set(
    xscale="log",
    yscale="linear",
#    ylim=[-0.03,1.05],
    xlabel="Separation, $r / (R_{NORM} s_0)$, -",
    ylabel="Reduction in $B(r) / B(R_{NORM} s_0)$",
)
sns.despine()
fig.tight_layout()
```

```python
fig, ax = plt.subplots()

wgrid = np.logspace(0.0, 1.5, 200)
plt.scatter(widths,b_norms)
ax.plot(wgrid, 0.5*(1 + 0.52 * wgrid / 7.9)**-2.0, linestyle="dashed")
ax.plot(wgrid, 0.5*(1 + 1.35 * wgrid / 7.9)**-1.0, linestyle="dashed", color = 'r')

ax.set(
    xlabel="s0",
    ylabel="max B",
    ylim=[0.0, None],
#    xscale="log",
)
```
