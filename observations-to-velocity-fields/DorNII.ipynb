```python
import sys
from pathlib import Path
import json
import numpy as np
from astropy.io import fits
#import strucfunc
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pickle
```

```python
dist = 50000 #parsecs
pix = 0.2 #arcsec
pc = dist*(2*np.pi) / (360 * 60 * 60)
s0=(0.9*pc)
pc,s0
```

```python
hdulist = fits.open("Obs/GAUS_NII6583.45_060_Will.fits")
```

```python
n = None
sb = hdulist[1].data[:n, :n].astype(np.float64)
vv = hdulist[2].data[:n, :n].astype(np.float64)
ss = hdulist[3].data[:n, :n].astype(np.float64)
```

```python
## Replace spurious values in the arrays
m = ~np.isfinite(sb*vv*ss) | (sb < 0.0)

m = m | (sb > 6e4) 
sb[m] = 0.0
vv[m] = np.nanmean(vv)
#ss[m] = 0.0
sb /= sb.max()

#good = (~m) & (sb > 0.001)
```

```python
fig, ax = plt.subplots()


dataRV=vv
plt.figure(1)
plt.imshow(dataRV, cmap='viridis')
cbar = plt.colorbar()
plt.clim(100,500)
cbar.set_label('km/s', rotation=270, labelpad=15)  

ax.set_xlabel('X')
ax.set_ylabel('Y')


ax.text(0.9, 0.1, '10 pc',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='black', fontsize=20)
    
plt.axhline(y=50, xmin=0.59, xmax=0.925, linewidth=2, color = 'k')


plt.gca().invert_yaxis()
```
