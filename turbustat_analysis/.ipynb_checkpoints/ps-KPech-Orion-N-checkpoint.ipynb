```python
from pathlib import Path
import numpy as np
import json
from astropy.io import fits
from astropy.utils.misc import JsonCustomEncoder
import astropy.units as u
import cmasher as cmr
from matplotlib import pyplot as plt
import turbustat.statistics as tss
import turbustat.simulator
from turbustat.simulator import make_3dfield
#from turb_utils import make_extended
import seaborn as sns
import sys

from turbustat.statistics import PowerSpectrum
```

```python
text_file_0 = open("path-data.txt", "r")
path_data = text_file_0.read()
```

```python
datapath_data = Path(path_data).expanduser()
```

```python
name = 'KPech-Orion-N'
```

```python
data = json.load(open(str(datapath_data) + '/' + name + "-m.json"))
sb = np.array(data["sb"])
vv = np.array(data["vv"])
#ss = np.array(data["ss"])
```

```python
 data['pix'] , data['pc']
```

```python
## Replace spurious values in the arrays
m = ~np.isfinite(sb*vv) | (sb < 0.0)

sb[m] = 0.0
vv[m] = np.nanmean(vv)
#ss[m] = 0.0
sb /= sb.max()

good = (~m) & (sb > 0.001)
```

```python
trim = (slice(0, 550), slice(50, 300))
```

```python
fig, ax = plt.subplots(figsize=(12, 12))


dataRV=vv[trim]

plt.figure(1)
plt.imshow(dataRV, cmap='RdBu_r')

cbar = plt.colorbar()
#plt.clim(200,350)
cbar.set_label('km/s', rotation=270, labelpad=15)  

ax.set_xlabel('X')
ax.set_ylabel('Y')


#ax.text(0.9, 0.1, '10 pc',
#        verticalalignment='bottom', horizontalalignment='right',
#        transform=ax.transAxes,http://localhost:8888/notebooks/Documents/Aeon/GitHub/PhD.Paper/power_spectrum/ps-KPech-Orion-N.ipynb#
#        color='black', fontsize=20)
    
#plt.axhline(y=50, xmin=0.59, xmax=0.925, linewidth=2, color = 'k')


plt.gca().invert_yaxis()
```

```python
r0 = 0.068 #pc
s0 = 0.002 #pc
k_r0 = (r0/( data['pix'] * data['pc']))**-1 #1/pix
k_s0 = (s0/( data['pix'] * data['pc']))**-1 #1/pix
```

```python
pspec = PowerSpectrum(fits.PrimaryHDU(vv[trim]), distance=440 * u.pc) 
pspec.run(verbose=True, xunit=u.pix**-1, low_cut=k_r0 / u.pix, high_cut=k_s0 / u.pix)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))


ax.scatter(pspec.freqs,pspec.ps1D)

ax.axvline(k_r0)
ax.axvline(k_s0)

ax.set(xscale='log', yscale='log', 
       xlabel='wavenumber $k$,1/pix',
       ylabel=r'$P(k),\ \mathrm{-}$'
      )
```
