```python
from pathlib import Path
import numpy as np
import json
from astropy.io import fits
from astropy.utils.misc import JsonCustomEncoder
import astropy.units as u
import cmasher as cmr
from matplotlib import pyplot as plt
import turbustat.statistics as tss
import turbustat.simulator
from turbustat.simulator import make_3dfield
#from turb_utils import make_extended
import seaborn as sns
import sys

from turbustat.statistics import PowerSpectrum
```

```python
text_file_0 = open("path-data.txt", "r")
path_data = text_file_0.read()
```

```python
datapath_data = Path(path_data).expanduser()
```

```python
name = 'KPech-Orion-H'
```

```python
data = json.load(open(str(datapath_data) + '/' + name + "-m.json"))
sb = np.array(data["sb"])
vv = np.array(data["vv"])
#ss = np.array(data["ss"])
```

```python
 data['pix'] , data['pc']
```

```python
## Replace spurious values in the arrays
m = ~np.isfinite(sb*vv) | (sb < 0.0)

sb[m] = 0.0
vv[m] = np.nanmean(vv)
#ss[m] = 0.0
sb /= sb.max()

good = (~m) & (sb > 0.001)
```

```python
trim = (slice(0, 513), slice(0, 350))
```

```python
plt.imshow(vv[trim])  
```

```python
r0 = 0.068 #pc
s0 = 0.002 #pc
k_r0 = (r0/( data['pix'] * data['pc']))**-1 #1/pix
k_s0 = (s0/( data['pix'] * data['pc']))**-1 #1/pix
```

```python
pspec = PowerSpectrum(fits.PrimaryHDU(vv[trim]), distance=440 * u.pc) 
pspec.run(verbose=True, xunit=u.pix**-1, low_cut=k_r0 / u.pix, high_cut=k_s0 / u.pix)
```

```python
plt.loglog(pspec.wavenumbers,pspec.ps1D)
```
